```python
# Export by key players overtime
years = ['1995', '1997', '2000', '2003', '2005', '2008', '2010', '2013', '2016']
CAN = []
FRA = []
ITA = []
RUS = []
GBR = []
JPN = []
DEU = []
USA = []
CHN = []
MEX = []
for y in years:
    file = 'downloads/country_partner_hsproduct4digit_year_' + y + '.csv'
    df_final = total_export(file)
    
    all_edges = df_final[['location_code','partner_code', 'export_portion']].values
    network = nx.DiGraph()
    network.add_weighted_edges_from(all_edges)
    eigen = eigenvector_centrality(network, weight = 'weight', max_iter = 1000)
    
#     df_total_export = df_final.groupby([df_final.location_code])['import_portion'].sum().reset_index(name = 'total_export')
    CAN.append(eigen['CAN'])
    FRA.append(eigen['FRA'])
    ITA.append(eigen['ITA'])
    RUS.append(eigen['RUS'])
    GBR.append(eigen['GBR'])
    JPN.append(eigen['JPN'])
    DEU.append(eigen['DEU'])
    USA.append(eigen['USA'])
    CHN.append(eigen['CHN'])
    MEX.append(eigen['MEX'])
```

```python
plt.rcParams["figure.figsize"] = (17,5)
fig = plt.figure()
ax = plt.subplot(111)
color_list = ["blue", "brown", "red", "black", "pink", "green", "purple", 'orange', 'grey', 'magenta']
x = np.arange(len(years))
plt.plot(x, CAN, label = 'Canada', linestyle = '--', c = color_list[0])
plt.plot(x, FRA, label = 'France', linestyle = '--', c = color_list[1])
plt.plot(x, ITA, label = 'Italy', linestyle = '--', c = color_list[2])
plt.plot(x, RUS, label = 'Russia', linestyle = '--', c = color_list[3])
plt.plot(x, GBR, label = 'United Kingdom', linestyle = '--', c = color_list[4])
plt.plot(x, JPN, label = 'Japan', linestyle = '--', c = color_list[5])
plt.plot(x, DEU, label = 'Germany', linestyle = '--', c = color_list[6])
plt.plot(x, USA, label = 'United States', linestyle = '--', c = color_list[7])
plt.plot(x, CHN, label = 'China', linestyle = '--', c = color_list[8])
plt.plot(x, MEX, label = 'Mexico', linestyle = '--', c = color_list[9])
plt.xticks(x, years)
plt.xlabel('Years')
plt.ylabel('Eigenvalue centrality')
plt.title('Eigenvalue centrality of key players overtime')

ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1),
          ncol=10, fancybox=True, shadow=True)

# plt.legend()
plt.show()
plt.show()
```
