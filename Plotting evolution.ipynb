```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
years = ['1995', '1997', '2000', '2003', '2005', '2008', '2010', '2013', '2016']
```

```python
density_graph = [0.47915881194755144,
 0.5458813301456206,
 0.5488424618859401,
 0.5731601731601732,
 0.5972533213912524,
 0.6345723242274967,
 0.6320346320346321,
 0.6158672917417959,
 0.595235497363157]
```

```python
def plot_evol(y, attribute):
    x = np.arange(len(years))
    plt.plot(x, y, 'k--')
    plt.scatter(x, y,c='k')
    plt.xticks(x, years)
    plt.xlabel('Years')
    plt.ylabel(attribute)
    plt.title(attribute+' overtime')
    plt.show()
```

```python
plot_evol(density_graph, 'Density')
```
